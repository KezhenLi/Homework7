```python
from numpy import zeros, empty, exp, linspace, array, copy
from pylab import plot, show, xlabel, ylabel
```

```python
# Constants
L = 1  # m
d = 0.1  # m
C = 1  # m/s
sigma = 0.3  # m
v = 100  # m/s
N = 1000  
a = L / N  
h = 10 ** -6  

def f(x):
    return C * (x * (L - x)) / L ** 2 * exp(-(x - d) ** 2 / (2 * sigma ** 2))

# Create arrays
x = linspace(0, L, N + 1)
phi0 = zeros(N + 1, float)
psi = array(f(x), float)
phi = []

# Main Loop
t = 0.0
c = h * v ** 2 / a ** 2
while t <= 0.01:
     phi.append(copy(phi0))
     phi0[1: N] += h * psi[1: N]
     psi[1: N] += c * (phi0[2: N + 1] + phi0[0: N - 1] - 2 * phi0[1: N])
     t += h

plot(x, phi[0])
plot(x, phi[10])
plot(x, phi[20])
plot(x, phi[30])
plot(x, phi[40])
plot(x, phi[50])
plot(x, phi[60])
plot(x, phi[70])
plot(x, phi[80])
plot(x, phi[90])
plot(x, phi[100])
xlabel("t(s)")
ylabel("phi(m)")
show()
```
